```python
# load libs
import pandas as pd
import numpy as np

# display floats with thousands separators and 0 decimal places
pd.options.display.float_format = '{:,.0f}'.format

# define number formats
fmtComma0 = '{:,.0f}'.format
fmtComma1 = '{:,.1f}'.format
fmtDollar = '${:,.0f}'.format
fmtPercent = '{:,.1f}%'.format

# set paths
strSch = 'Resources/schools_complete.csv'
strStu = 'Resources/students_complete.csv'

# create dataframes
dfSch = pd.read_csv(strSch)
dfStu = pd.read_csv(strStu)
```

```python
# create merged dataframe
dfMrg = pd.merge(dfStu, dfSch, how='left', on='school_name')
dfMrg
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# compute passing rates. I also prepended grade column's '9th' with a space char to assist crosstabbing later
dfMrg['reading_pass'] = dfMrg['reading_score'] > 69
dfMrg['math_pass'] = dfMrg['math_score'] > 69
dfMrg['grade'] = dfMrg['grade'].replace('9th', ' 9th')
dfMrg
```

```python
# generate District Summary dictionary, convert it to dataframe
dictDistSumm = {'Total Schools':        dfSch['school_name'].nunique(), \
                'Total Students':       f"{dfMrg['Student ID'].nunique():,.0f}", \
                'Total Budget':         f"${dfSch['budget'].sum():,.0f}", \
                'Average Math Score':   f"{dfMrg['math_score'].mean():,.1f}", \
                'Average Reading Score':f"{dfMrg['reading_score'].mean():,.1f}", \
                '% Passing Math':       f"{(dfMrg['math_pass'].sum() / dfMrg['Student ID'].nunique() * 100):,.1f}%", \
                '% Passing Reading':    f"{(dfMrg['reading_pass'].sum() / dfMrg['Student ID'].nunique() * 100):,.1f}%", \
                'Overall Passing Rate': f"{((dfMrg['reading_score'].mean() + dfMrg['math_score'].mean()) / 2):,.1f}%"}
dfDistSumm = pd.DataFrame(dictDistSumm, index=[0])
dfDistSumm
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# create School-level groupby
grpSchSumm = dfMrg.groupby(['school_name', 'type'])
# create School-level dataframe. drop unneeded cols & rename the rest
dfSchSumm = grpSchSumm.nunique()
dfSchSumm = dfSchSumm[['Student ID']]
dfSchSumm.index.names = ['School Name', 'School Type']
dfSchSumm = dfSchSumm.rename(columns = {'Student ID':'Total Students'})
dfSchSumm
```

```python
# append new cols
dfSchSumm['Total School Budget'] =   grpSchSumm['budget'].max()
dfSchSumm['Per Student Budget'] =    dfSchSumm['Total School Budget'] / dfSchSumm['Total Students']
dfSchSumm['Average Math Score'] =    grpSchSumm['math_score'].mean()
dfSchSumm['Average Reading Score'] = grpSchSumm['reading_score'].mean()
dfSchSumm['% Passing Math'] =        (grpSchSumm['math_pass'].sum() / grpSchSumm['Student ID'].nunique()) * 100
dfSchSumm['% Passing Reading'] =     (grpSchSumm['reading_pass'].sum() / grpSchSumm['Student ID'].nunique()) * 100
dfSchSumm['Overall Passing Rate'] =  (dfSchSumm['% Passing Math'] + dfSchSumm['% Passing Reading']) / 2
# create an unformatted copy of dfSchSumm for the 3 last 'Scores by...' exercises
dfSchSummRaw = dfSchSumm.copy(deep = True)
# format cols
dictFmts = {'Total Students': fmtComma0, 'Total School Budget':fmtDollar, 'Per Student Budget':fmtDollar, \
            'Average Math Score':fmtComma1, 'Average Reading Score':fmtComma1, '% Passing Math':fmtPercent, \
            '% Passing Reading':fmtPercent, 'Overall Passing Rate':fmtPercent}
for key, val in dictFmts.items():
    dfSchSumm[key] = dfSchSumm[key].apply(val)
dfSchSumm
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
# didn't define a Series; crosstab method did the trick
dfXTabRead = pd.crosstab(dfMrg['school_name'], dfMrg['grade'], values = dfMrg['reading_score'], aggfunc = 'mean')
# format as percents with one decimal place
for val in dfXTabRead.columns:
    dfXTabRead[val] = dfXTabRead[val].apply(fmtPercent)
dfXTabRead
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 630, 650, 1000000]
group_names = ['Per-Student Budgets 1-4 (<$585)', 'Per-Student Budgets 5-8 (<$630)', \
               'Per-Student Budgets 9-12 (<$650)', 'Per-Student Budgets 13-15 ($650+)']
```

```python
# get the unformatted dfSchSummRaw created previously. add cols for wtd avgs
dfSchSummRaw['Wtd Math Numerator'] = dfSchSummRaw['Total Students'] * dfSchSummRaw['Average Math Score']
dfSchSummRaw['Wtd Reading Numerator'] = dfSchSummRaw['Total Students'] * dfSchSummRaw['Average Reading Score']
dfSchSummRaw['Wtd Math Pass Numerator'] = dfSchSummRaw['Total Students'] * dfSchSummRaw['% Passing Math']
dfSchSummRaw['Wtd Reading Pass Numerator'] = dfSchSummRaw['Total Students'] * dfSchSummRaw['% Passing Reading']
dfSchSummRaw['Wtd Overall Pass Numerator'] = dfSchSummRaw['Total Students'] * dfSchSummRaw['Overall Passing Rate']
# bin it, group it
dfSchSummRaw['Budget Quartile'] = pd.cut(dfSchSummRaw['Per Student Budget'], spending_bins, labels=group_names)
grpSchSpend = dfSchSummRaw.groupby(['Budget Quartile'], as_index = False)
dfSchSpend = grpSchSpend.sum()
# drop unneeded cols from new dataframe
dfSchSpend = dfSchSpend[['Total Students', 'Wtd Math Numerator', 'Wtd Reading Numerator', \
                         'Wtd Math Pass Numerator', 'Wtd Reading Pass Numerator', 'Wtd Overall Pass Numerator']]
dfSchSpend
```

```python
# compute weighted average scores & percentages
dfSchSpend['Average Math Score'] = dfSchSpend['Wtd Math Numerator'] / dfSchSpend['Total Students']
dfSchSpend['Average Reading Score'] = dfSchSpend['Wtd Reading Numerator'] / dfSchSpend['Total Students']
dfSchSpend['% Passing Math'] = dfSchSpend['Wtd Math Pass Numerator'] / dfSchSpend['Total Students']
dfSchSpend['% Passing Reading'] = dfSchSpend['Wtd Reading Pass Numerator'] / dfSchSpend['Total Students']
dfSchSpend['Overall Passing Rate'] = dfSchSpend['Wtd Overall Pass Numerator'] / dfSchSpend['Total Students']
```

```python
# drop unneeded average-weighting cols
dfSchSpend = dfSchSpend[['Average Math Score', 'Average Reading Score', '% Passing Math', \
                         '% Passing Reading', 'Overall Passing Rate']]
```

```python
# define the column-formatter we'll use foor the last three exercises:
def formatCols(df):
    dictFmts = {'Average Math Score':fmtComma1, 'Average Reading Score':fmtComma1, \
                '% Passing Math':fmtPercent, '% Passing Reading':fmtPercent, 'Overall Passing Rate':fmtPercent}    
    for key, val in dictFmts.items():
        df[key] = df[key].apply(val)
    return
```

```python
formatCols(dfSchSpend)
dfSchSpend
```
